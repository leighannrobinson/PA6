```python
import math
```

```python
class calc():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
        
    def add(a,b):
        print("Addition: ",(a+b))
    def subtract(a,b):
        print("Subtraction: ",(a-b))
    def multiply(a,b):
        print("Multiplication: ",(a*b))
    def divide(a,b):
        print("Division: ",(a/b))
    def square(a):
        print("Squared: ",(a**2))
    def cube(a):
        print("Cubed: ",(a**3))
    def exponent(a,b):
        print("Exponent: ",(a**b))
    def sqrt(a):
        print("Square Root: ", math.sqrt(a))
    def log(a):
        print("Logarithm: ", math.log(a))
    def cos(a):
        print("Cosine: ", math.cos(a))
    def sin(a):
        print("Sine: ", math.sin(a))
    def tan(a):
        print("Tangent: ", math.tan(a))
    def acos(a):
        print("Arc Cosine: ", math.acos(a))
    def asin(a):
        print("Arc Sine: ", math.asin(a))
    def atan(a):
        print("Arc Tangent: ", math.atan(a))
```

```python
while True:
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")    
    print("5. Square")
    print("6. Cube")
    print("7. Exponent")
    print("8. Square Root")
    print("9. Logarithm")
    print("10. Cosine")
    print("11. Sine")
    print("12. Tangent")
    print("13. Arc Cosine")
    print("14. Arc Sin")
    print("15. Arc Tangent")
    print("16. Exit")
    
    choice=int(input("Which operation would you like to do? (1-16): "))
        
    if choice==1:
        a=int(input("Enter 1st number:"))
        b=int(input("Enter 2nd number:"))
        calc.add(a,b)

    elif choice==2:
        a=int(input("Enter 1st number:"))
        b=int(input("Enter 2nd number:"))
        calc.subtract(a,b)
    
    elif choice==3:
        a=int(input("Enter 1st number:"))
        b=int(input("Enter 2nd number:"))
        calc.multiply(a,b)
    
    elif choice==4:
        a=int(input("Enter 1st number:"))
        b=int(input("Enter 2nd number:"))
        if b == 0:
            print('Infinity')
        else:
            calc.divide(a,b)
    
    elif choice==5:
        a=int(input("Enter the number:"))
        calc.square(a)
    
    elif choice==6:
        a=int(input("Enter the number:"))
        calc.cube(a)

    elif choice==7:
        a=int(input("Enter 1st number:"))
        b=int(input("Enter 2nd number:"))
        calc.exponent(a,b)
          
    elif choice==8:
        a=int(input("Enter the number:"))
        calc.sqrt(a)
          
    elif choice==9:
        a=int(input("Enter the number:"))
        calc.log(a)
          
    elif choice==10:
        a=int(input("Enter the number:"))
        calc.cos(a)
          
    elif choice==11:
        a=int(input("Enter the number:"))
        calc.sin(a)
          
    elif choice==12:
        a=int(input("Enter the number:"))
        calc.tan(a)
          
    elif choice==13:
        a=int(input("Enter the number:"))
        calc.acos(a)
          
    elif choice==14:
        a=int(input("Enter the number:"))
        calc.asin(a)
          
    elif choice==15:
        a=int(input("Enter the number:"))
        calc.atan(a)
          
    elif choice==16:
        break
    else:
        print("Invalid Choice")
```
